```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dense(128)
conf1 = pd.read_csv("conf1.csv")
conf1.head()
```

```python
# Dense(128) + Dropout
conf2 = pd.read_csv("conf2.csv")
conf2.head()
```

```python
# Dense(128) + Dense(512)
conf3 = pd.read_csv("conf3.csv")
conf3.head()
```

```python
# Dense(128) + Dense(512) + Dropout
conf4 = pd.read_csv("conf4.csv")
conf4.head()
```

```python
df_valAcc = pd.DataFrame({
    'Conf1': conf1['val_acc'],
    'Conf2': conf2['val_acc'],
    'Conf3': conf3['val_acc'],
    'Conf4': conf4['val_acc']
})

df_valAcc.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valAcc, linewidth=2)
plt.legend(labels=['Dense(128)', 'Dense(128) with Dropout', 'Dense(512) + Dense(128)', 'Dense(512) + Dense(128) with Dropout'])
plt.title("Validation Accuracy of Multiple Configurations")
plt.savefig("442-valAcc", bbox_inches='tight')
plt.show()
```

```python
df_valLoss = pd.DataFrame({
    'Conf1': conf1['val_loss'],
    'Conf2': conf2['val_loss'],
    'Conf3': conf3['val_loss'],
    'Conf4': conf4['val_loss']
})

df_valLoss.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valLoss, linewidth=2)
plt.legend(labels=['Dense(128)', 'Dense(128) with Dropout', 'Dense(512) + Dense(128)', 'Dense(512) + Dense(128) with Dropout'])
plt.title("Validation Loss of Multiple Configurations")
plt.savefig("442-valLoss", bbox_inches='tight')
plt.show()
```
