```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rate0_0005 = pd.read_csv("rate0_0005.csv")
rate0_001 = pd.read_csv("../with_augmentation/with_augmentation.csv")
rate0_01 = pd.read_csv("rate0_01.csv")
```

```python
df_valAcc = pd.DataFrame({
    'rate0_0005': rate0_0005['val_acc'],
    'rate0_001': rate0_001['val_acc'],
    'rate0_01': rate0_01['val_acc']
})

df_valAcc.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valAcc, linewidth=2)
plt.legend(labels=['LR 0.0005', 'LR 0.001', 'LR 0.01'])
plt.title("Validation Accuracy of Multiple Configurations")
plt.savefig("445-valAcc", bbox_inches='tight')
plt.show()
```

```python
df_valLoss = pd.DataFrame({
    'rate0_0005': rate0_0005['val_loss'],
    'rate0_001': rate0_001['val_loss'],
    'rate0_01': rate0_01['val_loss']
})

df_valLoss.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valLoss, linewidth=2)
plt.legend(labels=['LR 0.0005', 'LR 0.001', 'LR 0.01'])
plt.title("Validation Loss of Multiple Configurations")
plt.savefig("445-valLoss", bbox_inches='tight')
plt.show()
```
