```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ep_100 = pd.read_csv("ep_100.csv")
ep_100_def = pd.read_csv("ep_100_def.csv")
```

```python
df_valAcc = pd.DataFrame({
    'ep_100': ep_100['val_acc'],
    'ep_100_def': ep_100_def['val_acc'],
})

df_valAcc.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valAcc, linewidth=2)
plt.legend(labels=['LR 0.0005', 'LR 0.001'])
plt.title("Validation Accuracy of Multiple Configurations")
plt.savefig("446-valAcc", bbox_inches='tight')
plt.show()
```

```python
df_valLoss = pd.DataFrame({
    'ep_100': ep_100['val_loss'],
    'ep_100_def': ep_100_def['val_loss'],
})

df_valLoss.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valLoss, linewidth=2)
plt.legend(labels=['LR 0.0005', 'LR 0.001'])
plt.title("Validation Loss of Multiple Configurations")
plt.savefig("446-valLoss", bbox_inches='tight')
plt.show()
```
