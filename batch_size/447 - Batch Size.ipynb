```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bs_16 = pd.read_csv("bs_16.csv")
bs_32 = pd.read_csv("bs_32.csv")
bs_64 = pd.read_csv("bs_64.csv")
```

```python
df_valAcc = pd.DataFrame({
    'bs_16': bs_16['val_acc'],
    'bs_32': bs_32['val_acc'],
    'bs_64': bs_64['val_acc']
})

df_valAcc.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valAcc, linewidth=2)
plt.legend(labels=['BS 16', 'BS 32', 'BS 64'])
plt.title("Validation Accuracy of Multiple Configurations")
plt.savefig("447-valAcc", bbox_inches='tight')
plt.show()
```

```python
df_valLoss = pd.DataFrame({
    'bs_16': bs_16['val_loss'],
    'bs_32': bs_32['val_loss'],
    'bs_64': bs_64['val_loss']
})

df_valLoss.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valLoss, linewidth=2)
plt.legend(labels=['BS 16', 'BS 32', 'BS 64'])
plt.title("Validation Loss of Multiple Configurations")
plt.savefig("447-valLoss", bbox_inches='tight')
plt.show()
```
