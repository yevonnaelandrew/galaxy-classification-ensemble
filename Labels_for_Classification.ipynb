```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("training_solutions_rev1.csv")
df.head()
```

```python
df = df[['GalaxyID', 'Class1.1', 'Class1.2', 'Class1.3', 'Class7.1', 'Class7.2', 'Class7.3', 'Class2.1', 'Class2.2', 'Class4.1', 'Class4.2']]
```

```python
df.head()
```

```python
len(df)
```

```python
df_completelySmooth = df[(df['Class1.1'] >= 0.469) & (df['Class7.1'] >= 0.50)]
len(df_completelySmooth)
```

```python
df_inbetweenSmooth = df[(df['Class1.1'] >= 0.469) & (df['Class7.2'] >= 0.50)]
len(df_inbetweenSmooth)
```

```python
df_cigarSmooth = df[(df['Class1.1'] >= 0.469) & (df['Class7.3'] >= 0.50)]
len(df_cigarSmooth)
```

```python
df_edgeOn = df[(df['Class1.2'] >= 0.430) & (df['Class2.1'] >= 0.602)]
len(df_edgeOn)
```

```python
df_spiral = df[(df['Class1.2'] >= 0.430) & (df['Class2.2'] >= 0.715) & (df['Class4.1'] >= 0.619)]
len(df_spiral)
```

```python
df_final = pd.concat([df_completelySmooth, df_inbetweenSmooth, df_cigarSmooth, df_edgeOn, df_spiral]).reset_index(drop=True)
df_final.head()
```

```python
df_final.describe().drop('GalaxyID', axis=1)
```

```python
df_final.to_csv("labels_classification.csv", index=False)
```
