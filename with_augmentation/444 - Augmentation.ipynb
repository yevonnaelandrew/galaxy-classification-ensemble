```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# With Augmentation
conf1 = pd.read_csv("with_augmentation.csv")
conf1.head()
```

```python
# Without Augmentation
conf2 = pd.read_csv("../top_layers/conf4.csv")
conf2.head()
```

```python
df_valAcc = pd.DataFrame({
    'Conf1': conf1['val_acc'],
    'Conf2': conf2['val_acc'],
})

df_valAcc.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valAcc, linewidth=2)
plt.legend(labels=['With Augmentation', 'Without Augmentation'])
plt.title("Validation Accuracy of Multiple Configurations")
plt.savefig("444-valAcc", bbox_inches='tight')
plt.show()
```

```python
df_valLoss = pd.DataFrame({
    'Conf1': conf1['val_loss'],
    'Conf2': conf2['val_loss'],
})

df_valLoss.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_valLoss, linewidth=2)
plt.legend(labels=['With Augmentation', 'Without Augmentation'])
plt.title("Validation Loss of Multiple Configurations")
plt.savefig("444-valLoss", bbox_inches='tight')
plt.show()
```
